```python
# استيراد المكتبات
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# إعدادات الرسومات داخل Jupyter
%matplotlib inline
sns.set(style="whitegrid")

# تحميل البيانات
file_path = "0b2b2865-3cad-4a65-974c-e9ac2e08879f.csv"  # أو ارفع الملف مباشرة عبر واجهة Jupyter
df = pd.read_csv(r"C:\Users\omarm\Downloads\archive (2)\Sample - Superstore.csv", encoding="ISO-8859-1")


# تحويل عمود التاريخ
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# (1) المبيعات حسب الفئة
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# (2) الربح مقابل الخصم
profit_discount = df[['Discount', 'Profit']].dropna()

# (3) أعلى 10 مدن من حيث المبيعات
top_cities_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

# رسم التحليلات
plt.figure(figsize=(15, 18))

# الرسم 1: المبيعات حسب الفئة
plt.subplot(3, 1, 1)
sns.barplot(x=sales_by_category.values, y=sales_by_category.index, palette="Blues_d")
plt.title("📊 إجمالي المبيعات حسب الفئة")
plt.xlabel("إجمالي المبيعات")
plt.ylabel("الفئة")

# الرسم 2: الربح مقابل الخصم
plt.subplot(3, 1, 2)
sns.scatterplot(data=profit_discount, x='Discount', y='Profit', color='purple')
plt.title("💰 الربح مقابل الخصم")
plt.xlabel("الخصم")
plt.ylabel("الربح")

# الرسم 3: أعلى 10 مدن من حيث المبيعات
plt.subplot(3, 1, 3)
sns.barplot(x=top_cities_sales.values, y=top_cities_sales.index, palette="Greens_d")
plt.title("🏙️ أعلى 10 مدن من حيث المبيعات")
plt.xlabel("إجمالي المبيعات")
plt.ylabel("المدينة")

plt.tight_layout()
plt.show()
```
